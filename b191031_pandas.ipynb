```python
import pandas as pd
df  = pd.read_csv("MELBOURNE_HOUSE_PRICES_LESS.csv")
df.head(10)
#df.tail(n) last no .of rows
```

```python
print(df.shape)
```

```python
df=df.dropna()#remove problematic rows
```

```python
print(df.shape)
```

```python
rooms=df["Rooms"]
print(rooms)
```

```python
price=list(df["Price"])
import matplotlib.pyplot as plt

print(min(price))
print(max(price))
print((max(price)-min(price))/10)

fig=plt.figure()

plt.hist(x=price,edgecolor="black",bins=40)
plt.xlabel("Price range")
plt.ylabel("No.of houses")
plt.title("Histogram of price")
plt.show()

rooms_count=df["Rooms"].value_counts()
rooms_count=dict(rooms_count)
rooms_count

a=list(rooms_count.keys())#room_size#better to convrt to list
b=list(rooms_count.values())#no_of_rooms

#Piechart
fig=plt.figure(figsize=(8,8))

plt.title("Pie Chart")
plt.pie(x=b,labels=a)

plt.legend()
plt.show()

#2.
reg_count=df["Regionname"].value_counts()
reg_count=dict(reg_count)
reg_count

c=list(reg_count.keys())#room_size
d=list(reg_count.values())#no_of_rooms

#Piechart
fig=plt.figure(figsize=(8,8))

plt.title("Pie Chart")
plt.pie(x=d,labels=c)

plt.legend()
plt.show()

#4.
typ=list(df["Type"])

""""print(min(price))
print(max(price))
print((max(price)-min(price))/10)"""

fig=plt.figure()

plt.hist(x=typ,edgecolor="green",bins=5)
plt.xlabel("Type")
plt.ylabel("No.of houses")
plt.title("Histogram of Type")
plt.show()
```

```python
#Type
typ_count=df["Type"].value_counts()
typ_count=dict(typ_count)
typ_count
```

```python
#Suburb
sub_count=df["Suburb"].value_counts()
sub_count=dict(sub_count)
sub_count
```

```python
#Bar graph for Type
k=list(typ_count.keys())
l=list(typ_count.values())

fig=plt.figure(figsize=(5,5))
plt.bar(x=k,height=l)
plt.xlabel("Type")
plt.ylabel("No of Houses")
plt.title("Bar graph for Type")
plt.show()
```

```python
#Scatter plot for Price and distance
price=list(df["Price"])
dist=list(df["Distance"])

fig=plt.figure(figsize=(7,7))
#line plot>>plt.plot()
plt.xlabel("distance")
plt.ylabel("price")
plt.plot(dist,price,linestyle="",marker="o",markeredgecolor="red",markerfacecolor="black")

plt.show()
```
